```python
import numpy as np
import cv2
from PIL import Image
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

1.1.1 Image Color Order with different libraries

```python
fig = plt.figure(figsize=(12, 5))

# setting values to rows and column variables
rows = 2
columns = 4

img = cv2.imread("D:\Image processing\lab\pic\dolly.png")

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


B = (img[ : , : , 2])
G = (img[ : , : , 1])
R = (img[ : , : , 0])


# BGR
fig.add_subplot(rows, columns, 1)

plt.imshow(img)
plt.title("BGR")

fig.add_subplot(rows, columns, 2)

plt.imshow(B, cmap='gray')
plt.title("B")

fig.add_subplot(rows, columns, 3)

plt.imshow(G, cmap='gray')
plt.title("G")

fig.add_subplot(rows, columns, 4)

plt.imshow(R, cmap='gray')
plt.title("R")

# RGB
fig.add_subplot(rows, columns, 5)

plt.imshow(RGB_img)
plt.title("RGB")

fig.add_subplot(rows, columns, 6)

plt.imshow(R, cmap='gray')
plt.title("R")

fig.add_subplot(rows, columns, 7)

plt.imshow(G, cmap='gray')
plt.title("G")

fig.add_subplot(rows, columns, 8)

plt.imshow(B, cmap='gray')
plt.title("B")
```

1.1.2 Image Reshape

```python
fig = plt.figure(figsize=(12, 5))
rows = 1
columns = 4
shapeimg = RGB_img.shape

tpimg = np.transpose(RGB_img)
shapetpimg = tpimg.shape

aximg = np.moveaxis(RGB_img,-1,0)
shapeaximg = aximg.shape

rsimg = np.reshape(RGB_img, (3,415,-1))
shapersimg = rsimg.shape



fig.add_subplot(rows, columns, 1)
plt.imshow(RGB_img[:,:,0],cmap='gray')
plt.title('X = (H, W, CH)\nX = {}\nR_Original'.format(shapeimg))
# print('')

fig.add_subplot(rows, columns, 2)
plt.imshow(tpimg[0,:,:],cmap='gray')
plt.title('X = (CH, W, H)\nX = {}\nR_Tranpose'.format(shapetpimg))
# print('')

fig.add_subplot(rows, columns, 3)
plt.imshow(aximg[0,:,:],cmap='gray')
plt.title('X = (CH, H, W)\nX = {}\nR_Moveaxis'.format(shapeaximg))
# print('')

fig.add_subplot(rows, columns, 4)
plt.imshow(rsimg[0,:,:],cmap='gray')
plt.title('X = (CH, H, W)\nX = {}\nR_Reshape'.format(shapersimg))
print('')




```
